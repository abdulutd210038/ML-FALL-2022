## Dataset description

335 landslides (including rockslides, rock avalanches, soil slides, and slides) in the Zagros mountain belt of southwest Iran using a digital elevation model (DEM) with 30 m resolution, Google EarthTM images, and field investigation. Individual landslides have volumes ranging from 10**4 (the lower limit of resolution) to 3 × 10**10 m**3 and cover surface areas ranging from 10**3 to 10**8 m**2. The relationship between landslide volume (VL) and area (AL) is well described by a power law of the form VL = AL**alpha, where alpha = 1.49 over five orders of magnitude of AL and seven orders of magnitude of VL. Frequency-size (i.e., volume) distribution is heavy tailed, following a power law for the largest landslides (>10**7 m**3) with a scaling exponent beta = 1.51. Non-power-law behavior for smaller landslides is probably an artifact due to the relatively low resolution, such that the dataset is essentially missing many small landslides. Comparison of these results with the other published data sets around the globe shows that the Zagros landslides are relatively larger, and they show similar scaling behavior to those observed in other regions, especially with regard to data sets where landslides are deep-seated or relatively large and occur in relatively resistant materials (e.g., consolidated rocks, as opposed to soil).

### Querstion 1: 

Apply one hot encoder to the dataset on relevent columns.

Answer here: 

```python
# It should be noted that OneHotEncoder is suitable for column with a few categorical variables. So, I applied this Encoder technique on columns where we have maximum of two categorical variables.
# OneHotEncoder is Applied to the Area(Region) column
df2['Area (Region)'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
enc.fit(df2[['Area (Region)']])
result = enc.transform(df2[['Area (Region)']])
ohe_columns = ['Area (Region)'+"_"+str(x) for x in enc.categories_[0]]
result_train = pd.DataFrame(result, columns=ohe_columns)
df3 = pd.concat([df2, result_train], axis=1)
df3


```

```python
# We will now apply OneHotEncoder to another column. 
df3['Surf descr (Constructed surface)'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
enc.fit(df3[['Surf descr (Constructed surface)']])
result = enc.transform(df3[['Surf descr (Constructed surface)']])
ohe_columns = ['Surf descr (Constructed surface)'+"_"+str(x) for x in enc.categories_[0]]
result_train = pd.DataFrame(result, columns=ohe_columns)
df4 = pd.concat([df3, result_train], axis=1)
df4
```

### Querstion 3: 
In our dataset which columns can be encoded with label encoding?
Is it good idea to use this encoding methos?

Answer here: 

```python
# Those columns in which a categorical variables are involved can be encoded with the Label Encoding. So, in our dataset columns such as Area_(Region), Slide (Slide type) etc can be encoded with the Label encoding. It is good idea to do label encoding when we have a  larger number of categorical variables. For smaller number of categorical variables we prefer OneHotEncoding Method, because if OneHotEncoding is performed on a larger data, then it would create a very large distribution of data. Hence, Label Encoding has advantage in this case. However, one drawback of the Label Encoding is that it allows Machine to itself rank the data by looking at 0,1,2,3. An example of Label Encoding is given below.
```

```python
df5['Slide (Slide type)'].unique()
```

```python
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels
label_encoder = preprocessing.LabelEncoder()
  
df5['Slide (Slide type)']= label_encoder.fit_transform(df5['Slide (Slide type)'])
  
df5['Slide (Slide type)'].unique()
```
