2. Plot distribution of the earthquakes and change color and size of the epicenters based on age and magnitude. like the following example:
####   


```python
import seaborn as sns
sns.set_theme(style="whitegrid")


cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True,)
g = sns.relplot(
    data=df,
    x="longitude", y="latitude",
    hue="year", size="mag",
    palette=cmap, sizes=(100, 200), height = 10, aspect = 1.25
)
g.set(xscale="linear", yscale="linear")
g.ax.xaxis.grid(True, "minor", linewidth=1.25)
g.ax.yaxis.grid(True, "minor", linewidth=1.25)
g.despine(left=True, bottom=True)

```
